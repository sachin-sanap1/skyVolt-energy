```python
df['wind_speed_category'] = df['wind_speed_category'].fillna("Na")
df['wind_direction_category'] = df['wind_direction_category'].fillna("Na")
```

```python
df['plant_installation_date'] = pd.to_datetime(df['plant_installation_date'])
```

```python
cols = df[['wind_speed', 'wind_direction', 'generator_speed', 'theoretical_power', 'power_output']]
for col in cols:
    Q3 = df[col].quantile(0.75)
    Q1 = df[col].quantile(0.25)
    IQR = Q3 - Q1 
    upper_limit = Q3 + 1.5 * IQR
    lower_limit = Q1 - 1.5 * IQR
    outliers = df[(df[col]>upper_limit) | (df[col]<lower_limit)]
    print(f"Outlier in {col} : {len(outliers)}")
```

```python
plt.figure()
df['installation_year'].value_counts().sort_index().plot(kind='pie',autopct='%1.1f%%')
plt.title('Plants Installed Per Year')
plt.show()
```

```python
plt.figure()
df['wind_speed_category'].value_counts().plot(kind='bar')
plt.xlabel('Wind Speed Category')
plt.ylabel('Count')
plt.title('Wind Speed Category Distribution')
plt.show()
```

```python
plt.figure()
df['state'].value_counts().plot(kind='bar')
plt.xlabel('state')
plt.ylabel('Number of Plants')
plt.grid()
plt.title('State-wise Plant Distribution')
plt.show()
```

```python
plt.figure()
df['wind_direction_category'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.title('Wind Direction Category Share')
plt.show()
```

```python
plt.figure()
plt.scatter(df['wind_speed'], df['power_output'])
plt.xlabel('Wind Speed')
plt.ylabel('Power Output')
plt.title('Wind Speed vs Power Output')
plt.show()
```

```python
plt.figure()
df.groupby('wind_speed_category')['power_output'].mean().plot(kind='bar')
plt.xlabel('Wind Speed Category')
plt.ylabel('Average Power Output')
plt.title('Average Power Output by Wind Speed Category')
plt.show()
```

```python
pivot_df = df.pivot_table(
    values='power_output',
    index='state',
    columns='sector',
    aggfunc='mean'
)

pivot_df.plot(kind='bar')
plt.xlabel('State')
plt.ylabel('Average Power Output')
plt.title('Average Power Output by State and Sector')
plt.show()
```
